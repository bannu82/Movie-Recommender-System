```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(30)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits,on='title')
movies.info()
```

```python
# genres 
# id	
# keywords
# title
# overview          
# cast                  
# crew                  
```

```python
movies = movies[['id'  , 'title', 'keywords','genres'  ,'overview' ,'cast' ,'crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj) :
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords']= movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj) :
        if counter !=3 :
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
movies['cast']= movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
movies.iloc[1].crew
```

```python
def fetchDirector(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj) :
        if i['job']  == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetchDirector)
```

```python
movies.head(3)
```

```python
movies.iloc[1:4].overview    
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(3)
```

```python
# Sam Worthington
# SamWorthington
```

```python
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" " , "") for i in x])
movies['overview'] = movies['overview'].apply(lambda x : [i.replace(" " , "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" " , "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" " , "") for i in x])
```

```python
movies.head(5)
```

```python
movies['tag'] = movies['genres'] + movies['overview'] + movies['cast'] + movies['crew']
```

```python
movies.head(2)
```

```python
new_df = movies[['id' , 'title' , 'tag']] 
```

```python
new_df.head()
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:" " . join(x))
new_df['tag'] = new_df['tag'].apply(lambda x:x.lower())
```

```python
new_df.head(1)
```

```python
import nltk 
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y =[]

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000 , stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
cv.get_feature_names_out()
```

```python

# calculating distance of each vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
# sorted(list(enumerate(similarity[1])), reverse=True , key= lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movies_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movies_index]
    movies_list = sorted(list(enumerate(distance)), reverse=True , key= lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
    return
```

```python
recommend("Iron Man")
```
